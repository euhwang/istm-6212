## Aggregate Functions and Subqueries

The following statement indicates that the same facility can have multiple nature of business.
The first cell shows that there are 2,615 distinct facilities, but 2,660 distinct facility and nature of business combinations.

```python
%%sql
select count(*), 'Facility' as Attributes_Compared from 
(
select facility_name
from foodinspect
group by facility_name
)t1

UNION 

select count(*), 'Facility | Nature of Business' as Attributes_Compared from 
(
select facility_name, nature_of_business
from foodinspect
group by facility_name, nature_of_business
)t2
;
```

The first cell lists out all the different facility name and nature of business combinations. The second cell identifies the businesses that have multiple nature of business.

Using the data output below, one can see the following:
- Auntie Annes is considered both a restaurant and a 'rest/mur'
- Baskin Robbins is considered both a restaurant and ice cream
- Boba Tea Company is considered both a drink based establishment and a tea shop
- etc.

```python
%%sql
select count(*), facility_name, nature_of_business
from foodinspect
group by facility_name, nature_of_business
order by facility_name asc;
```

```python
%%sql
select count(*), t1.facility_name
from 
(
select facility_name, nature_of_business
from foodinspect
group by facility_name, nature_of_business
)t1
group by t1.facility_name
having count(*)>1
order by t1.facility_name asc
;
```

The following statement indicates that the same owner can have multiple keys. The first cell shows that there are 1,852 distinct owners, but 1,892 distinct owner name and owner key combinations. The second cell identifies the businesses that have multiple nature of business. As you can see below, Burger King has 4 different owner key, thus a owner key is not unique for each owner.
```python
%%sql
select count(*), 'Owner' as Attributes_Compared from 
(
select owner_name
from foodinspect
group by owner_name
)t1

UNION ALL

select count(*), 'Owner | Owner Key' as Attributes_Compared from 
(
select owner_name, owner_key
from foodinspect
group by owner_name, owner_key
)t2
;
```

The following statement checks to see if the code and description have an one-to-one relationship. As you can see below Action Code/Description and Result Code/Description are normalized. However, Violation Code is not unique. Violation Code '04  70' has two descriptions (LACK OF REFRIGERATION and LACK OF REFRIGERATIONS). Ideally, there should be a look up table for the violation code, so we only have to update the look up table. However, in case, we will need to update Violation Code '04  70' to either 'LACK OF REFRIGERATION' or 'LACK OF REFRIGERATIONS'. For this exercise, I will updated it to 'LACK OF REFRIGERATION' for consistency.

```python
%%sql
select count(*), 'Action Code' as Attributes_Compared from 
(
select action_code
from foodinspect
group by action_code
)t1

UNION ALL

select count(*), 'Action Code | Action Description' as Attributes_Compared from 
(
select action_code, action_desc
from foodinspect
group by action_code, action_desc
)t2

UNION ALL

select count(*), 'Result Code' as Attributes_Compared from 
(
select result_code
from foodinspect
group by result_code
)t1

UNION ALL

select count(*), 'Result Code | Result Description' as Attributes_Compared from 
(
select result_code, result_desc
from foodinspect
group by result_code, result_desc
)t2

UNION ALL

select count(*), 'Violation Code' as Attributes_Compared from 
(
select violation_code
from foodinspect
group by violation_code
)t1

UNION ALL

select count(*), 'Violation Code | Violation Description' as Attributes_Compared from 
(
select violation_code, violation_desc
from foodinspect
group by violation_code, violation_desc
)t2
;
```

```python
%%sql
select count(*), t1.violation_code
from 
(
select violation_code, violation_desc
from foodinspect
group by violation_code, violation_desc
)t1
group by t1.violation_code
having count(*)>1
order by t1.violation_code asc
;
```

```python
%%sql
select distinct violation_code, violation_desc
from foodinspect
where violation_code = '04  70'
order by violation_code
;
```

## Transformations

By adding year, month, and day attributes to the table, we can easily perform analysis on scenarios such as which year, month, day had the most inspections. The following cells alters the table and adds additonal attributes and updates the attributes with values.

```python
%%sql
SELECT TO_CHAR(inspection_date, 'YYYY-MM-DD') AS inspection_day_of_year, 
    TO_CHAR(inspection_date, 'YYYY') AS inspection_year,
    TO_CHAR(inspection_date, 'MM') AS inspection_month,
    TO_CHAR(inspection_date, 'DD') AS inspection_day,
    TO_CHAR(inspection_date, 'Day') AS inspection_day_of_week_str,
    TO_CHAR(inspection_date, 'D') AS inspection_day_of_week,
    CASE WHEN CAST(TO_CHAR(inspection_date, 'D') AS INTEGER) >= 6 
        THEN 1 
        ELSE 0
    END AS inspection_is_weekend,
    CASE WHEN CAST(TO_CHAR(inspection_date, 'D') AS INTEGER) < 6 
        THEN 1 
        ELSE 0
    END AS inspection_is_weekday,
    TO_CHAR(inspection_date, 'HH24') AS inspection_hour_24,
    TO_CHAR(inspection_date, 'Q') AS inspection_quarter
FROM foodinspect
LIMIT 10
;
```

```python
%%sql
ALTER TABLE foodinspect
ADD COLUMN inspection_day_of_year char(10),
ADD COLUMN inspection_year INTEGER,
ADD COLUMN inspection_month INTEGER,
ADD COLUMN inspection_day INTEGER
;
```

```python
%%sql
UPDATE foodinspect SET (inspection_day_of_year, inspection_year, inspection_month, inspection_day) = 
(        TO_CHAR(inspection_date, 'YYYY-MM-DD'), 
    CAST(TO_CHAR(inspection_date, 'YYYY') AS INTEGER),
    CAST(TO_CHAR(inspection_date, 'MM') AS INTEGER),
    CAST(TO_CHAR(inspection_date, 'DD') AS INTEGER)
 )
;
```

```python
%%sql
SELECT inspection_date,
       inspection_day_of_year, 
       inspection_year, 
       inspection_month, 
       inspection_day
FROM foodinspect
LIMIT 10
;
```

```python
%%sql
select count(*), 'Program Category' as Attributes_Compared from 
(
select program_category
from foodinspect
group by program_category
)t1

UNION ALL

select count(*), 'Program Category | Program Category Description' as Attributes_Compared from 
(
select program_category, program_category_description
from foodinspect
group by program_category, program_category_description
)t2
;
```

The following cells drops the table program_category_lk if it exist, then creates the table, and inserts the distinct values for program category/description into the newly created table. Lastly, the last command checks to make sure all the records have been inserted correctly.

```python
%%sql
DROP TABLE IF EXISTS program_category_lk;
CREATE TABLE program_category_lk 
(
    PROGRAM_CATEGORY CHAR(4),
    PROGRAM_CATEGORY_DESCRIPTION VARCHAR(40)
)
```

```python
%%sql
INSERT INTO program_category_lk (PROGRAM_CATEGORY, PROGRAM_CATEGORY_DESCRIPTION)
SELECT distinct PROGRAM_CATEGORY, PROGRAM_CATEGORY_DESCRIPTION FROM foodinspect;
```

```python
%%sql
select * from program_category_lk;
```

The following cell is an example of what you can do with a look up table. Ideally, you do not need to have the program category description in the main table. The main table can join to the look up table to get the descriptions. 

If, however, the program category description was not normalized, you can create a similar lookup table with normalized description data and leave the original values in the main table. Then, re-insert the description text in a new column joining on the program category. 

```python
%%sql
select fi.facility_name,
       fi.program_category,
       pc.program_category_description
from foodinspect fi join program_category_lk pc on fi.program_category = pc.program_category
limit 10
;
```

```python
%%sql
ALTER TABLE foodinspect
ADD COLUMN new_pc_desc VARCHAR(40)
;
```

```python
%%sql
UPDATE foodinspect AS a
SET new_pc_desc = b.program_category_description
FROM program_category_lk AS b
WHERE a.program_category = b.program_category;
```

One reason to keep the original progam category description may be that you want to keep it for historical purposes. However, for the purpose to normalizing data, I would create a separate look up table for the program category and only keep the program category (code) in the main table.
